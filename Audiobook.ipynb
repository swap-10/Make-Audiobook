```python
import torch
from scipy.io.wavfile import write
tacotron2 = torch.hub.load('NVIDIA/DeepLearningExamples:torchhub', 'nvidia_tacotron2', model_math='fp16')
tacotron2 = tacotron2.to('cuda')
tacotron2.eval()
waveglow = torch.hub.load('NVIDIA/DeepLearningExamples:torchhub', 'nvidia_waveglow', model_math='fp16')
waveglow = waveglow.remove_weightnorm(waveglow)
waveglow = waveglow.to('cuda')
waveglow.eval()
```

```python
allowedchars = [chr(i) for i in range(ord('\u0020'), ord('\u007f'))]
print(allowedchars)
```

```python
f = open('library.txt', 'r')
text = f.read()
for character in text:
  if character not in allowedchars:
    text = text.replace(character, '')

print(text)
print("Len in chars:", len(text))
print("Len in words:", len(text.split()))
```

```python
def finite_split(strings, length):
  retlist = [strings[0+i:i+length] for i in range(0, len(strings), length)]
  retlist = [[' '.join(split)] for split in retlist]
  return retlist, len(retlist)
test = ["aohfioehw", "fwhofhoiwb", "fuwbfkjjabg", "kabhlighuwefla", "sogoiejg", "sijviojwijv", "shvuhwinvc", "ancuhiwuv"]
test = finite_split(test, 2)
print(test)
test = ["aohfioehw", "fwhofhoiwb", "fuwbfkjjabg", "kabhlighuwefla", "sogoiejg", "sijviojwijv", "shvuhwinvc", "ancuhiwuv"]
test[0:1] = "a"
print(test)
```

```python
text = text.replace("[", "(")
text = text.replace("]", ")")
text = text.split('.')
print(text[:2])
index = 0
while index < len(text):
  num = 1
  sentence = text[index]
  if (len(sentence.split()) > 10):
    text[index:index+1], num = finite_split(sentence.split(), 10)
  index = index + num

print(text)
for item in text:
  if type(item) is not type([]):
    text.remove(item)
  if item in [' ', '', '  ']:
    text.remove(item)
print(text)
```

```python
text = text[:200]  # Keeping shorter for demonstration
```

```python
utils = torch.hub.load('NVIDIA/DeepLearningExamples:torchhub', 'nvidia_tts_utils')
counter=0
for i in text:
  i = str(i)
  print(i)
  sequences, lengths = utils.prepare_input_sequence([i])
  with torch.no_grad():
      mel, _, _ = tacotron2.infer(sequences, lengths)
      audio = waveglow.infer(mel)
  audio_numpy = audio[0].data.cpu().numpy()
  rate = 22050
  write("/content/AudioFiles/audio%04d.mp3" %counter, rate, audio_numpy)
  with open('/content/filenames.txt', 'w') as fn:
    fn.write("file 'AudioFiles/audio%04d.mp3'\n" %(counter))
  print("This done", counter)
  counter+=1
```
